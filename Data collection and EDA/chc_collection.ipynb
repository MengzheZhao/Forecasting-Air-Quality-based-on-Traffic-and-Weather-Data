## Air quality

```python
station1 = pd.read_csv('../778project/data/airquality/christchurch/riccartonroad_h.csv')
station1
```

```python
station1['DateTime'] = pd.to_datetime(station1['DateTime'], format='%d/%m/%Y %I:%M:%S %p').dt.strftime('%Y-%m-%d')
station1.dtypes
```

```python
station1 = station1[(station1['DateTime']>='2020-01-01') & (station1['DateTime']<='2023-12-31')]
station1.rename(columns={'DateTime': 'Date'}, inplace=True)
station1['Date'] = pd.to_datetime(station1['Date'])
station1
```

```python
station2 = pd.read_csv('../778project/data/airquality/christchurch/woolston_h.csv')
station2
```

```python
station2['DateTime'] = pd.to_datetime(station2['DateTime'], format='%d/%m/%Y %I:%M:%S %p').dt.strftime('%Y-%m-%d')
station2 = station2[(station2['DateTime']>='2020-01-01') & (station2['DateTime']<='2023-12-31')]
station2.rename(columns={'DateTime': 'Date'}, inplace=True)
station2['Date'] = pd.to_datetime(station2['Date'])
station2
```

### CO

```python
full_range = pd.date_range(start='2020-01-01', end='2023-12-31', freq='D')
full_range = pd.DataFrame({'Date':full_range})
full_range
```

```python
co_1 = station1[['Date', 'CO (mg/m3)']]
co_1 = co_1.groupby('Date')['CO (mg/m3)'].mean().reset_index()
co_1
```

```python
co_2 = station2[['Date', 'CO (mg/m3)']]
co_2 = co_2.groupby('Date')['CO (mg/m3)'].mean().reset_index()
co_2
```

```python
co_df = full_range.merge(co_1, on='Date', how='left')
co_df = co_df.merge(co_2, on='Date', how='left')
co_df
```

```python
co_df['Daily Average CO (mg/m3)'] = co_df[['CO (mg/m3)_x', 'CO (mg/m3)_y']].mean(axis=1)
co_df = co_df[['Date', 'Daily Average CO (mg/m3)']]
co_df
```

### NO1

```python
no1_df = station1[['Date', 'NO (ug/m3)']]
no1_df.rename(columns={'NO (ug/m3)': 'Daily Average NO (ug/m3)'}, inplace=True)
no1_df = no1_df.groupby('Date')['Daily Average NO (ug/m3)'].mean().reset_index()
no1_df
```

### NO2

```python
no2_df = station1[['Date', 'NO2 (ug/m3)']]
no2_df.rename(columns={'NO2 (ug/m3)': 'Daily Average NO2 (ug/m3)'}, inplace=True)
no2_df = no2_df.groupby('Date')['Daily Average NO2 (ug/m3)'].mean().reset_index()
no2_df
```

### SO2

```python
so2_df = station2[['Date', 'SO2 (ug/m3)']]
so2_df.rename(columns={'SO2 (ug/m3)': 'Daily Average SO2 (ug/m3)'}, inplace=True)
so2_df = so2_df.groupby('Date')['Daily Average SO2 (ug/m3)'].mean().reset_index()
so2_df
```

```python
so2_df = full_range.merge(so2_df, on='Date', how='left')
so2_df
```

### PM2.5

```python
pm25_df = station2[['Date', 'PM2.5 (ug/m3)']]
pm25_df.rename(columns={'PM2.5 (ug/m3)': 'Daily Average PM2.5 (ug/m3)'}, inplace=True)
pm25_df = pm25_df.groupby('Date')['Daily Average PM2.5 (ug/m3)'].mean().reset_index()
pm25_df = full_range.merge(pm25_df, on='Date', how='left')
pm25_df
```

### PM10

```python
pm10_df = station2[['Date', 'PM10 (ug/m3)']]
pm10_df.rename(columns={'PM10 (ug/m3)': 'Daily Average PM10 (ug/m3)'}, inplace=True)
pm10_df = pm10_df.groupby('Date')['Daily Average PM10 (ug/m3)'].mean().reset_index()
pm10_df = full_range.merge(pm10_df, on='Date', how='left')
pm10_df
```

### Air quality

```python
chc_df = full_range.merge(co_df, on='Date', how='left')
chc_df = chc_df.merge(no1_df, on='Date', how='left')
chc_df = chc_df.merge(no2_df, on='Date', how='left')
chc_df = chc_df.merge(so2_df, on='Date', how='left')
chc_df = chc_df.merge(pm25_df, on='Date', how='left')
chc_df = chc_df.merge(pm10_df, on='Date', how='left')
chc_df
```

## Weather

```python
weather = pd.read_csv('../778project/data/weather/christchurch_weather.csv', skiprows=10)
weather = weather[:-6]
weather
```

```python
weather['Day(Local_Date)'] = weather['Day(Local_Date)'].astype(str).str.split('.').str[0]
weather['Day(Local_Date)'] = pd.to_datetime(weather['Day(Local_Date)'], format='%Y%m%d')
weather = weather.rename(columns={'Day(Local_Date)':'Date'})
weather = weather[(weather['Date']>='2020-01-01') & (weather['Date']<='2023-12-31')]
weather = weather.drop(columns=['Station'])
weather
```

## Traffic

```python
tms = pd.read_csv('../778project/data/traffic/TMS_daily_traffic_counts_API.csv')
tms
```

```python
tms = tms[tms['regionName']=='11 - Canterbury']
tms
```

```python
tms['startDate'] = pd.to_datetime(tms['startDate'], dayfirst=True)
tms['startDate'] = tms['startDate'].dt.date
tms['startDate'] = pd.to_datetime(tms['startDate'])
tms.dtypes
```

```python
tms_light = tms[tms['classWeight']=='Light']
tms_light = tms_light[(tms_light['startDate']>='2020-01-01') & (tms_light['startDate']<='2023-12-31')]
tms_light
```

```python
tms_light = tms_light.groupby('startDate')['trafficCount'].mean().reset_index()
tms_light
```

```python
tms_heavy = tms[tms['classWeight']=='Heavy']
tms_heavy = tms_heavy[(tms_heavy['startDate']>='2020-01-01') & (tms_heavy['startDate']<='2023-12-31')]
tms_heavy = tms_heavy.groupby('startDate')['trafficCount'].mean().reset_index()
tms_heavy
```

```python
tms_light = tms_light.rename(columns={'startDate':'Date', 'trafficCount':'lightCount'})
tms_heavy = tms_heavy.rename(columns={'startDate':'Date', 'trafficCount':'heavyCount'})
tms_light
```
