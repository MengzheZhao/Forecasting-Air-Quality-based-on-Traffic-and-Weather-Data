### NO2

```python
no2 = pd.read_csv('../778project/data/airquality/wellington/no2_h.csv')
no2
```

```python
no2.rename(columns={'Date (NZST)':'Date', ' Nitrogen Dioxide 1hr Average (ug/m3) (ug/m3)':'Daily Average NO2 (ug/m3)'}, inplace=True)
no2['Date'] = pd.to_datetime(no2['Date'], format='%d/%m/%Y %H:%M:%S').dt.strftime('%Y-%m-%d')
no2['Date'] = pd.to_datetime(no2['Date'])
no2 = no2.groupby('Date')['Daily Average NO2 (ug/m3)'].mean().reset_index()
no2 = no2[(no2['Date']>='2020-01-01') & (no2['Date']<='2023-12-31')]
no2
```

```python
full_range = pd.date_range(start='2020-01-01', end='2023-12-31', freq='D')
full_range = pd.DataFrame({'Date':full_range})
full_range
```

```python
no2_df = full_range.merge(no2, on='Date', how='left')
no2_df
```

## Weather

```python
weather = pd.read_csv('../778project/data/weather/wellington_weather.csv', skiprows=10)
weather = weather[:-6]
weather
```

```python
weather['Day(Local_Date)'] = weather['Day(Local_Date)'].astype(str).str.split('.').str[0]
weather['Day(Local_Date)'] = pd.to_datetime(weather['Day(Local_Date)'], format='%Y%m%d')
weather = weather.rename(columns={'Day(Local_Date)':'Date'})
weather = weather[(weather['Date']>='2020-01-01') & (weather['Date']<='2023-12-31')]
weather = weather.drop(columns=['Station'])
weather
```

## Traffic

```python
tms = pd.read_csv('../778project/data/traffic/TMS_daily_traffic_counts_API.csv')
tms
```

```python
tms = tms[tms['regionName']=='09 - Wellington']
tms
```

```python
tms['startDate'] = pd.to_datetime(tms['startDate'], dayfirst=True)
tms['startDate'] = tms['startDate'].dt.date
tms['startDate'] = pd.to_datetime(tms['startDate'])
tms.dtypes
```

```python
tms_light = tms[tms['classWeight']=='Light']
tms_light = tms_light[(tms_light['startDate']>='2020-01-01') & (tms_light['startDate']<='2023-12-31')]
tms_light
```

```python
tms_light = tms_light.groupby('startDate')['trafficCount'].mean().reset_index()
tms_light
```

```python
tms_heavy = tms[tms['classWeight']=='Heavy']
tms_heavy = tms_heavy[(tms_heavy['startDate']>='2020-01-01') & (tms_heavy['startDate']<='2023-12-31')]
tms_heavy = tms_heavy.groupby('startDate')['trafficCount'].mean().reset_index()
tms_heavy
```

```python
tms_light = tms_light.rename(columns={'startDate':'Date', 'trafficCount':'lightCount'})
tms_heavy = tms_heavy.rename(columns={'startDate':'Date', 'trafficCount':'heavyCount'})
tms_light
```
